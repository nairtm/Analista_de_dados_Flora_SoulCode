```python
'''
Exercício sobre medidas de estatística descritiva para variáveis quantitativas

Considere a base de dados a seguir:
https://docs.google.com/spreadsheets/d/1q64TPaErzCYgXX6Exp1hopq_L2JnRT3j/edit?usp=sharing&ouid=110283574113665030533&rtpof=true&sd=true

Você foi contratado para fazer as medidas de estatística descritivas para as variáveis quantitativas de uma base de dados sobre possíveis motivos de desgaste dos funcionários de uma empresa que os levaram a pedir demissão, sua tarefa é calcular essas medidas  e trazer possíveis causas a partir da interpretação dos resultados.

Dica: Antes de calcular as medidas separe os dados em dois grupos, os que pediram demissão e os que não pediram

Obs.: Faremos a correção do trabalho na parte da manhã da aula de quarta-feira

'''
```

```python
import pandas as pd
import numpy as np
import statistics as st
```

```python
# importando uma tabela a partir de um link da internet
df = pd.read_csv('https://docs.google.com/spreadsheets/d/1A6kuRtAsrVxsvhxVwFLpRVwZGqBRsYkR/export?format=csv')
display(df)
```

```python
# Dividir o demitidos e não demitidos
demissao = df[df['Demissão'] == 'Sim']
nao_demissao = df[df['Demissão'] == 'Não']
display(demissao)
```

```python
display(nao_demissao)
```

```python
# Escolhe as variáveis quantitativas
i = df.select_dtypes(include=[np.number]).columns #selecionar as colunas de um DataFrame que contêm dados numéricos

# Calcula as valores descritivos
def valores_descritivos(valor):
    valores = {
        'Média': valor.mean(),
        'Mediana': valor.median(),
        'Moda': valor.mode().iloc[0],
        'Desvio Padrão': valor.std(),
        'Valor Mínimo': valor.min(),
        'Valor Máximo': valor.max(),
        'Variância': valor.var(),
        '1° Quartil': valor.quantile(0.25),
        '2° Quartil': valor.quantile(0.50),
        '3° Quartil': valor.quantile(0.75),
        'Amplitude Interquartilica': valor.quantile(0.75) - valor.quantile(0.25),
        'Limite Inferior': valor.quantile(0.25) - 1.5 * (valor.quantile(0.75) - valor.quantile(0.25)),
        'Limite Superior': valor.quantile(0.75) + 1.5 * (valor.quantile(0.75) - valor.quantile(0.25))
    }

    return pd.DataFrame(valores)

# Separa os valores em demitidos e não demitidos
valores_demissao = valores_descritivos(demissao[i])
valores_nao_demissao = valores_descritivos(nao_demissao[i])

# DEMITIDOS
print("Variáveis quantitativas para quem teve demissão:")
print(valores_demissao)
```

```python
# NÃO DEMITIDOS
print("Variáveis quantitativas para quem NÃO teve demissão:")
print(valores_nao_demissao)
```

```python
# Calcular a correlação entre as colunas
correlacao_demissao = np.corrcoef(demissao[i].T)
correlacao_nao_demissao = np.corrcoef(nao_demissao[i].T)


correlacao_demissao_df = pd.DataFrame(correlacao_demissao, index=i, columns=i)
correlacao_nao_demissao_df = pd.DataFrame(correlacao_nao_demissao, index=i, columns=i)


print("Correlação para quem teve demissão:")
print(correlacao_demissao_df)
```

Interpretaçao das correlações:

3.   Nível hierárquico e Total de anos trabalhados: 0.742761.Quanto mais tempo o funcionário tem de anos trabalhados mais ele tem chances de subir na hierarquia da empresa.
4.   Total de Anos Trabalhados e Anos Trabalhados na Empresa: 0.767402. Indicando que muitos funcionários que pediram demissão tinham uma parte significativa de sua experiência de trabalho acumulada na própria empresa. Isso sugere uma fidelidade ao empregador ao longo do tempo, mas também pode revelar alguns possíveis motivos de insatisfação e desgaste que os levaram a pedir demissão, como estagnação.
5.   Total de anos trabalhados na empresa e Anos trabalhados na função atual: 0.810889. A falta de oportunidades de promoção ou mudança de função, contribui para o desgaste e a insatisfação, o que pode levar o funcionário pedir demissão.
6.   Anos trabalhados na função atual e Anos trabalhando com o atual gestor: 0.781087. O funcionário permanece muito tempo com o mesmo gestor não há rotatividade de liderança, isto pode causar um desgaste.

8.   Nível Hierárquico e Total de Anos Trabalhados na Empresa: 0.614977. Indica que nem todos os funcionários que têm muitos anos de serviço são necessariamente promovidos a posições superiores. Isso pode levar a um sentimento de estagnação e frustração entre aqueles que esperam ser promovidos.
9.   Total de Anos Trabalhados e Renda Mensal: 0.742277.Funcionários com mais anos de experiência tendem a ter salários mais altos.
10.   Total de Anos Trabalhados e Anos Trabalhados na Função Atual: 0.621348.Isso implica que muitos funcionários tendem a passar bastante tempo em suas funções atuais.
11.   Anos Desde a Última Promoção e Total de Anos Trabalhados na Empresa: 0.702128. Isso significa que, em muitos casos, funcionários que têm um longo período de serviço na empresa também passaram muitos anos desde a sua última promoção.
12.   Anos Desde a Última Promoção e Anos Trabalhando com o Atual Gestor: 0.683218. A correlação positiva indica que o tempo que um funcionário passa trabalhando com o mesmo gestor está associado ao tempo desde sua última promoção. Isso sugere que funcionários que estão há muito tempo sob a supervisão do mesmo gestor também tendem a passar longos períodos sem serem promovidos.
