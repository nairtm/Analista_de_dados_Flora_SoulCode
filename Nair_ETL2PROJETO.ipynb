```python
import pandas as pd
```

```python
df = pd.read_csv('https://storage.googleapis.com/projeto1nair/dados-brutos/dado.csv')
display(df)
```

```python
#armazenando as informações da coluna e transformando em uma lista
nome = df['nome']
nome= list(nome)
print(nome)
idade = df['idade']
idade= list(idade)
print(idade)
salario = df['salário']
salario= list(salario)
print(salario)
cargo = df['cargo']
cargo= list(cargo)
print(cargo)
departamento= df['departamento']
departamento= list(departamento)
print(departamento)
```

```python
# convertendo para string
nome_str = [str(n) for n in nome]
idade_str = [str(i) for i in idade]
salario_str = [str(s) for s in salario]
cargo_str = [str(c) for c in cargo]
departamento_str = [str(d) for d in departamento]

print("Nomes:", nome_str)
print("Idades:", idade_str)
print("Salários:", salario_str)
print("Cargos:", cargo_str)
print("Departamentos:", departamento_str)
```

```python
indices_para_remover = []

#removendo os dads de erros, desconhecido e "vazios"
for x in range(len(nome_str)):
    if (idade_str[x] == 'erro' or idade_str[x] == 'nan' or idade_str[x] == 'Desconhecido' or
        salario_str[x] == 'erro' or salario_str[x] == 'nan' or salario_str[x] == 'Desconhecido' or
        cargo_str[x] == 'nan' or cargo_str[x] == 'erro' or cargo_str[x] == 'Desconhecido' or
        departamento_str[x] == 'nan' or departamento_str[x] == 'erro' or departamento_str[x] == 'Desconhecido'):

        indices_para_remover.append(x)


for j in sorted(indices_para_remover, reverse=True):
    nome_str.pop(j)
    idade_str.pop(j)
    salario_str.pop(j)
    cargo_str.pop(j)
    departamento_str.pop(j)


print("Nomes:", nome_str)
print("Idades:", idade_str)
print("Salários:", salario_str)
print("Cargos:", cargo_str)
print("Departamentos:", departamento_str)
```

```python
df_oficial = pd.DataFrame({'nome': nome_str,'idade': idade_str,'salário': salario_str,'cargo': cargo_str,'departamento': departamento_str})
display(df_oficial)
df_oficial.to_csv('tratados.csv', index=False)
```
