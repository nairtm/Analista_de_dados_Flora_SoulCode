```python
import pandas as pd
```

```python
df = pd.read_csv('https://storage.googleapis.com/aula6nair/dados-brutos/dados%20-%20P%C3%A1gina1.csv')
print (df)
```

```python
display(df)
```

```python
df.info()
```

```python
'''
Verificamos que ao extrair os dados, os mesmos vieram limpos e tipados
corretamente, porém, desejamos fazer uma transformação criando uma coluna que
calcula automaticamente os valores de imc.
'''
df['imc'] = df['peso'] / (df['altura'] ** 2)
display(df)
```

```python
imcs = df['imc']
imcs =  list(imcs)
for i in range(len(imcs)):
  imcs[i] = round(imcs[i], 2)
df['imc'] = imcs
display(df)
```

```python
df.to_csv('tratados.csv', index=False)
#não vem os indices index = False
#gera o arquivo tratados
```
